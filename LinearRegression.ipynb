```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Dados de exemplo
X = np.array([
    [1, 4],
    [2, 5],
    [3, 5],
    [4, 6],
    [5, 6],
    [6, 7],
    [7, 7],
    [8, 8],
    [9, 8],
    [10, 9]
])

y = np.array([50, 55, 60, 63, 68, 72, 75, 80, 85, 90])
```

```python
#Modelo
model = LinearRegression()
model.fit(X, y)

#Previsões
y_pred = model.predict(X)
```

```python
#Resultados
print("Coeficientes (a1 e a2):", model.coef_)
print("Intercepto (b):", model.intercept_)
print("Equação: y = {:.2f}*x1 + {:.2f}*x2 + {:.2f}".format(model.coef_[0], model.coef_[1], model.intercept_))
```

```python
#Visualizar
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], y, color='blue', label='Dados (x1)')
plt.plot(X[:, 0], y_pred, color='red', label='Previsão do modelo')
plt.xlabel('Horas de estudo (x1)')
plt.ylabel('Nota no exame')
plt.legend()
plt.show()
```
